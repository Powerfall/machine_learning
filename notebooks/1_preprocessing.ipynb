```python
import pandas as pd
```

```python
pokemon = pd.read_csv("../data/Pokemon.csv", index_col=0)
pokemon.head()
```

```python
pokemon.info()
```

```python
pokemon.isnull().sum()
```

Приводим бинарные признаки к 1 или 0

```python
mapping = {False : 0,
           True : 1}

pokemon['Legendary'] = pokemon['Legendary'].map(mapping)
pokemon
```

Перекодируем категориальные признаки. Для каждого типа сделаем собственный префикс.

```python
tmp1 = pokemon[['Name','Type 1']]
tmp2 = pokemon[['Name', 'Type 2']]
types1 = pd.get_dummies(tmp1, prefix='t', columns=['Type 1'])
types2 = pd.get_dummies(tmp2, prefix='t', columns=['Type 2'])
ccat = pd.concat([types1, types2], join="inner")
allTypes = ccat.groupby(['Name']).sum()
res = pd.DataFrame(pokemon).merge(allTypes, on='Name', how='inner')
res = res.drop(columns=['Type 1', 'Type 2'])
res.info()
res.head()
```

```python
res.to_csv('../data/Pokemon_preprocessed.csv')
```
