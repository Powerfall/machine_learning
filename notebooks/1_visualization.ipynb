```python
# будем отображать графики прямо в jupyter'e
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#графики в svg выглядят более четкими
%config InlineBackend.figure_format = 'svg' 

#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
```

```python
pokemon = pd.read_csv("../data/Pokemon_preprocessed.csv", index_col=0)
pokemon.head()
```

На графиках ниже можно увидеть какие характеристики преобладали в зависимости от их поколения.

```python
pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].plot()
```

```python
pokemon.groupby('Generation').mean()[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']].plot(kind='bar', rot=0)
```

Ниже представлены графики отношений каждой характеристики к каждой.

```python
cols = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
sns_plot = sns.pairplot(pokemon[cols])
```

Ниже представлен график корреляции между аттрибутами.

```python
plt.figure(figsize=(10,6))
sns.heatmap(pokemon.corr(), annot=True)
plt.show()
```

Ниже представлен график, показывающий какие типы покемонов преобладают

```python
plt.title('Count Plot')
plt.xticks(rotation = 45)
sns.countplot(pokemon['Type 1'])
sns.countplot(pokemon['Type 2'])
plt.xlabel('Types')
```

Ниже представлен график, на котором видно насколько характеристики легендарных покемонов превосходят характеристики обычных.

```python
sns.boxplot(y="Total", x="Legendary", data=pokemon)
```
